```python
## Created a function to change age int/float dtype
## column into categorial column object dtype for more understanding

def fun(age):
    if age<=13:
        return "kid"
    elif age>13 and age<=22:
        return "teenager"
    elif 22<age<=40:
        return 'Adult'
    elif 40<age<=51:
        return 'mid age'
    else:
        return "Mature"
```

```python
df['cat_age'] = df['age'].apply(fun)
```
